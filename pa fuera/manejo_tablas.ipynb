```python
import pandas as pd

# Cargar datos desde el archivo Excel
df = pd.read_excel('costo_produccion.xlsx')

# Transponer el DataFrame (filas a columnas)
df_transpuesto = df.transpose()

# Obtener una serie con todos los elementos de las columnas en orden
columna_unica = pd.Series(df_transpuesto.values.flatten(), name='Columna_Unica')

# Guardar la serie en un nuevo archivo Excel
columna_unica.to_excel('costo_produccion_apilados.xlsx', index=False)

print("Resultados guardados en 'costo_produccion_apilados.xlsx'")
```

```python
import pandas as pd

# Crear una lista con los cacopios a repetir
cacopios = ["Guamal", "Pivijay", "Nueva granada"]

# Repetir cada nombre 36 veces
cacopios_repetidos = [nombre for nombre in cacopios for _ in range(132)]

# Crear un DataFrame con una sola columna
df_resultado = pd.DataFrame(cacopios_repetidos, columns=['cacopios_Repetidos'])

# Guardar el DataFrame en un nuevo archivo Excel
df_resultado.to_excel('cacopios_repetidos.xlsx', index=False)

print("cacopios repetidos guardados en 'cacopios_repetidos.xlsx'")
```
